#Hypothesis Testing
## Null Hypothesis
-Theres no significant relationship between a movie's genre , release timing, budget or audience rating/ reviews and box office revenue.

## Alternative Hypothesis
-Theres is a significant relationship between a movie's genre , release timing, budget or audience rating/ reviews and box office revenue.



```python
# Extract the relevant columns
x = movies['numvotes']
y = movies['domestic_gross']

# Pearson correlation coefficient
rho = np.corrcoef(x, y)[0][1]

# Standard deviations for calculating the slope
s_y = y.std()
s_x = x.std()

# Calculating the slope of the regression line
m = rho * (s_y / s_x)

# Means of the variables x and y
mean_y = y.mean()
mean_x = x.mean()

# Calculate the intercept of the regression line
c = mean_y - m * mean_x

# Print the regression line
print(f'y = {m}x + {c}')

# Visualize the regression line
import matplotlib.pyplot as plt

# Scatter plot of the data points
plt.figure(figsize=(12, 6))
plt.scatter(x, y, color='skyblue', alpha=0.5, label='Data Points')

# Plot the regression line
regression_line = m * x + c
plt.plot(x, regression_line, color='red', label='Regression Line')

plt.xlabel('Number of Votes')
plt.ylabel('Domestic Gross')
plt.title('Linear Regression: Domestic Gross vs. Number of Votes')
plt.legend()
plt.show()
```

The scatter plot visualizes the relationship between the number of votes a movie receives (x-axis) and its domestic gross revenue (y-axis). The red line represents the linear regression line, which shows the overall trend of the relationship.

```python
# Selecting independent variables (features)
X = movies[['numvotes', 'averagerating', 'runtime_minutes']]  # Example: numvotes, averagerating, runtime_minutes
y = movies['domestic_gross']  # Dependent variable

# Add a constant to the independent variables (for the intercept)
X_scaled = sm.add_constant(X)

# Fit the model using statsmodels
model = sm.OLS(y, X_scaled).fit()

# Print the regression summary
print(model.summary())
```

We carried out a multiple linear regression analysis to examine the relationship between a movie's domestic gross revenue (domestic_gross) and three independent variables: numvotes, averagerating, and runtime_minutes.

#Model Fit:

Explains about 44.8% of variance in domestic gross (R-squared).
Statistically significant overall (F-statistic).

#Predictors:

Number of votes: Positive and significant impact on domestic gross.
Average rating: Negative and significant impact (requires further investigation).
Runtime: Negative and significant impact.

#Diagnostics:

Residuals not normally distributed (potential for model improvement).

In summary, the model suggests that the number of votes is positively associated with domestic gross, while average rating and runtime are negatively associated. However, the non-normality of residuals and the negative coefficient for average rating may indicate further investigation or model adjustments are necessary.

# Multiple Linear Regression

```python
# Extract relevant columns
X = movies[['averagerating', 'genres', 'release_year']]
y = movies['domestic_gross']


movies_shuffled = movies.sample(frac=1, random_state=42)
X = movies_shuffled[['averagerating', 'genres', 'release_year']]
y = movies_shuffled['domestic_gross']

#Determine the split ratio
split_ratio = 0.8
split_index = int(len(movies_shuffled) * split_ratio)


#Split the data
train_data = movies_shuffled[:split_index]
test_data = movies_shuffled[split_index:]

#Create 'primary genre' for both train and test data
train_data['primary_genre'] = train_data['genres'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else 'Unknown')
test_data['primary_genre'] = test_data['genres'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else 'Unknown')

#create domestic_gross_millions
train_data['domestic_gross_millions'] = np.log1p(train_data['domestic_gross'])/1000000
test_data['domestic_gross_millions'] = np.log1p(test_data['domestic_gross'])/1000000 # Create the column in test_data as well

#then separate the parts for training and testing
X_train = train_data[['averagerating', 'primary_genre', 'release_year']]
y_train = train_data['domestic_gross_millions']
x_test = test_data[['averagerating', 'primary_genre', 'release_year']]
y_test = test_data['domestic_gross_millions']


#print the sizes
print("Training set size:", len(X_train))
print("Testing set size:", len(x_test))

```

We first do Data splitting: Divides the dataset into training (80%) and testing (20%) sets using a random split for model evaluation.

Feature engineering:
Creates a 'primary_genre' column by extracting the first genre listed for each movie to simplify genre representation.

Data separation: Splits both training and testing data into independent (features: averagerating, primary_genre, release_year) and dependent (target: domestic_gross_millions) variables.

```python
#Make the model and compute coeffients
movies['primary_genre'] = movies['genres'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else 'Unknown')

#Filter out release years with single observations
year_counts = movies['year'].value_counts()
valid_years = year_counts[year_counts > 1].index
movies = movies[movies['year'].isin(valid_years)]

#convert release_year to categorical after filtering
movies['release_year'] = pd.Categorical(movies['year'])

movies['domestic_gross_millions'] = np.log1p(movies['domestic_gross'])/1000000
regression_formula = "domestic_gross_millions ~ averagerating + C(primary_genre) + release_year"
model = smf.ols(regression_formula, data=pd.concat([train_data, test_data])).fit()
print(model.summary())
```

Here we  build and train an Ordinary Least Squares (OLS) regression model to predict domestic_gross_millions (likely a transformed version of domestic gross revenue) using averagerating, primary_genre, and release_year as predictors.

```python
# Identify outliers in domestic_gross
movies.boxplot(column=['domestic_gross'])

# Remove outliers (optional)
movies = movies[movies['domestic_gross'] < movies['domestic_gross'].quantile(0.95)]
```
